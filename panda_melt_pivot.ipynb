## 과제 내용

해당 데이터프레임은 컬럼 중 변수명_연도로 된 것을 다음과 같이 정리하고자 한다.

연도 정보가 없는 컬럼은 df_county로 만든다.

연도 정보가 있는 컬럼들은 df_ym으로 만든다.



1-1 VF_indcom_counties_Q222.csv를 읽어 df 변수에 저장

1-2 마지막행은 지운다 .

1-3 cfips를 정수로 바꾼다음, 문자타입으로 변환한다.

```python
df = pd.read_csv('VF_indcom_counties_Q222.csv')
df.drop(index=df.index[-1], inplace=True)
df['cfips'] = df['cfips'].astype('int32').astype(str)
```

```python
df.head()
```


2-1 df의 컬럼에서 a-z가 3자와 숫자 2자리로 끝나는 컬럼명을 추출하여 cols_ym에 저장한다.

hint: str.contains 의 regular expression을 사용하라.  

정규 표현식 참조: [a-z]{n}: a-z의 글자 n자 $ 문자열의 끝

2-2 이를 제외한 컬럼을 cols_notym으로 저장한다.

```python
cols_ym = df.columns[df.columns.str.contains('[a-z]{3}[0-9]{2}$')]
cols_ym
```

```python
cols_notym = df.columns[~df.columns.str.contains('[a-z]{3}[0-9]{2}$')]
cols_notym
```

3-1 cols_notym 으로 선택한 데이터프레임을  df_county로 저장하고, 해당 필드 중에서 **cfips**를 제외하고 df에서 없앤다.

3-2 df_county에서 cfips를 인덱스로 한다.

```python
df_county = df[cols_notym]
df.drop(columns=cols_notym[1:], inplace=True)
df_county.set_index('cfips', inplace=True)
```

```python
df_county.head()
```

4.cols_ym 은  변수명_년월로 구성되었다.  

df에서 cols_yn의 컬럼들을 변수로 만들어 cfips, cols_ym의 컬럼 값, cols_ym의 값으로 된 3개의 컬럼을 지닌 데이터프레임을 df_ym로 만든다.

hint: DataFrame의 melt 함수를 사용하여 만든다.

```python
df_ym = df.melt(id_vars='cfips', value_vars=cols_ym)
df_ym.head()
```

5-1 variable 컬럼의 변수 내용 문자열을 뒤에서 첫 번째로 '_' 문자가 나온 지점을 기준으로 두 개의 문자로 나누어 첫 번째는 'var_name'으로 두 번째는 'ym'으로  컬럼명으로 하여  이를 df_ym에 추가한다.

예) orders_rank_aug19 -> orders_rank, aug19 

hint: str.rpartition을 이용하면 간단

*이를 이용하지 않고도 해보자

5-2 variable 컬럼 제거.



```python
# str.rpartition을 이용할 경우
df_ym[['var_name', 'ym']] = df_ym['variable'].str.rpartition('_')[[0, 2]]
```

```python
# str.rpartition을 사용하지 않고 map을 이용
df_ym['tmp'] = df_ym['variable'].map(lambda x: x.split('_'))
df_ym['var_name'] = df_ym['tmp'].map(lambda x: "_".join(x[:-1]))
df_ym['ym'] = df_ym.pop('tmp').map(lambda x: x[-1])
```

```python
df_ym.drop(columns='variable', inplace=True)
```

6-1 df_ym의 ym 컬럼을 뒤의 숫자를 떼어 year 컬럼으로 만들고 앞의 문자열은 month 컬럼으로 만든다음 ym을 삭제한다.

hint: str.extract 메소드 와 정규표현식 활용

6-2 month의 내용을 숫자 월로 바꾼다. 다음 dict를 참조한다.

{'apr': 4, 'aug': 8, 'dec': 12, 'feb': 2, 'jan': 1, 'jul': 7, 'jun': 6, 'mar': 3, 'may': 5, 'nov': 11, 'oct': 10, 'sep': 9}

hint: replace 메소드 

6-3 year에 각각 2000을 더한다.

6-4. 'ym'을 삭제한다.

```python
df_ym[['month', 'year']] = df_ym['ym'].str.extract("^([a-z]+)([0-9]+)$")
```

```python
df_ym['month'] = df_ym['month'].replace({'apr': 4, 'aug': 8, 'dec': 12, 'feb': 2, 'jan': 1, 'jul': 7, 'jun': 6, 'mar': 3, 'may': 5, 'nov': 11, 'oct': 10, 'sep': 9})
df_ym['year'] = df_ym['year'].astype('int32') + 2000
df_ym.drop(columns='ym', inplace=True)
```

7-1 var_name의 unique한 요소를 체크한다.

hint: 제대로 되었다면 'orders_rank', 'merchants_rank', 'gmv_rank', 'top3industries',
       'avg_traffic'

7-2 cfips, year, month는 index로 하고, var_name은 컬럼으로 value는 컬럼의 값으로 피벗을 수행한다. 

DataFrame pivot을 받드시 사용한다. 중복된 행에 의해서 에러가 발생한다면, 해당 행중 index가 작은 값을 취하고 나머지는 버린다.

hint. pivot_table 을 이용하여 두 개 이상 들어간 값을 확인한다.
index가 작은 값을 취하는 것은 groupby와 head를 이용한다.
그 다음 중복되는 것을 지우고, 취했던 행을 포함시킨다.



```python
df_ym['var_name'].unique()
```

```python
df_ym.pivot(index=['cfips', 'year', 'month'], columns='var_name', values='value')
```

```python
df_cnt = df_ym.pivot_table(index=['cfips', 'year', 'month'], columns=df_ym['var_name'], values='value', aggfunc='count')
df_cnt.loc[(df_cnt > 1).any(axis=1)]
```

```python
df_cnt.loc[(df_cnt > 1).any(axis=1)].index.get_level_values(0).unique()
```

```python
tmp = df_ym.query("cfips in ['47125', '51013']").groupby(['cfips', 'var_name', 'year', 'month']).head(1)
```

```python
df_ym.drop(df_ym.query("cfips in ['47125', '51013']").index, inplace=True)
df_ym = pd.concat([df_ym, tmp], axis=0)
```

```python
df_cnt = df_ym.pivot_table(index=['cfips', 'year', 'month'], columns=df_ym['var_name'], values='value', aggfunc='count')
df_cnt.loc[(df_cnt > 1).any(axis=1)]
```
